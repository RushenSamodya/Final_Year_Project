1. ARIMA Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import os
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('seaborn-v0_8')
sns.set(rc={'figure.figsize':(15, 8)})

# Function to check stationarity using Augmented Dickey-Fuller test
def check_stationarity(series, window=12):
    """
    Check if a time series is stationary using ADF test and rolling statistics
    """
    # Rolling statistics
    rolling_mean = series.rolling(window=window).mean()
    rolling_std = series.rolling(window=window).std()
    
    # Plot rolling statistics
    plt.figure(figsize=(12, 6))
    plt.plot(series, label='Original')
    plt.plot(rolling_mean, label='Rolling Mean')
    plt.plot(rolling_std, label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.tight_layout()
    plt.show()
    
    # ADF Test
    print('Results of Augmented Dickey-Fuller Test:')
    adf_test = adfuller(series.dropna(), autolag='AIC')
    adf_output = pd.Series(adf_test[0:4], index=['ADF Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in adf_test[4].items():
        adf_output['Critical Value (%s)' % key] = value
    print(adf_output)
    
    # Interpret results
    if adf_test[1] <= 0.05:
        print("The series is stationary (reject the null hypothesis)")
    else:
        print("The series is not stationary (fail to reject the null hypothesis)")
    
    return adf_test[1] <= 0.05  # Return True if stationary

# Function to perform a grid search for finding the best ARIMA parameters
def find_best_arima_parameters(series, max_p=3, max_d=2, max_q=3):
    """
    Grid search to find best ARIMA parameters
    """
    best_aic = float('inf')
    best_params = None
    
    for p in range(max_p + 1):
        for d in range(max_d + 1):
            for q in range(max_q + 1):
                try:
                    model = ARIMA(series, order=(p, d, q))
                    results = model.fit()
                    aic = results.aic
                    
                    if aic < best_aic:
                        best_aic = aic
                        best_params = (p, d, q)
                        
                    print(f"ARIMA({p},{d},{q}) - AIC: {aic}")
                except:
                    continue
    
    print(f"\nBest ARIMA parameters: {best_params} with AIC: {best_aic}")
    return best_params

# Function to build and evaluate ARIMA model
def build_arima_model(train, test, order):
    """
    Build ARIMA model, make predictions and evaluate
    """
    # Fit the model
    model = ARIMA(train, order=order)
    model_fit = model.fit()
    
    # Print model summary
    print(model_fit.summary())
    
    # Forecast
    forecast = model_fit.forecast(steps=len(test))
    
    # Convert forecast to Series for easier plotting
    forecast_series = pd.Series(forecast, index=test.index)
    
    # Plot actual vs predicted
    plt.figure(figsize=(15, 6))
    plt.plot(train[-30:], label='Training Data (last 30 points)')
    plt.plot(test, label='Actual')
    plt.plot(forecast_series, label='Forecast')
    plt.title(f'ARIMA{order} - Forecast vs Actual')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Make sure both test and forecast don't have NaN values
    # Create masks for non-NaN values in both series
    valid_idx = ~(np.isnan(test) | np.isnan(forecast))
    
    # Check if we have enough valid points for evaluation
    if sum(valid_idx) < 2:
        print("WARNING: Not enough valid data points for evaluation. Metrics may be unreliable.")
        print(f"Valid data points: {sum(valid_idx)} out of {len(test)}")
        # Return some placeholder metrics if we can't calculate them
        return model_fit, forecast_series, {'rmse': np.nan, 'mae': np.nan, 'r2': np.nan}
    
    # Filter out NaN values
    test_clean = test[valid_idx]
    forecast_clean = forecast[valid_idx]
    
    # Calculate metrics using cleaned data
    try:
        rmse = np.sqrt(mean_squared_error(test_clean, forecast_clean))
        mae = mean_absolute_error(test_clean, forecast_clean)
        r2 = r2_score(test_clean, forecast_clean)
        
        print(f'RMSE: {rmse:.4f}')
        print(f'MAE: {mae:.4f}')
        print(f'R: {r2:.4f}')
    except Exception as e:
        print(f"Error calculating metrics: {str(e)}")
        return model_fit, forecast_series, {'rmse': np.nan, 'mae': np.nan, 'r2': np.nan}
    
    return model_fit, forecast_series, {'rmse': rmse, 'mae': mae, 'r2': r2}

# Function to process and analyze a water level station data
def analyze_water_level_station(file_path, station_name):
    """
    Process, analyze and build ARIMA model for water level station data
    """
    print(f"\n{'='*80}")
    print(f"Analyzing station: {station_name}")
    print(f"{'='*80}")
    
    # Read data
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"Error reading file {file_path}: {str(e)}")
        return None
    
    # Check if data exists
    if df.empty:
        print(f"No data found for {station_name}")
        return None
    
    # Parse dates and set as index
    try:
        df.columns = ['datetime', 'unit', 'water_level', 'label']
        df['datetime'] = pd.to_datetime(df['datetime'])
        df.set_index('datetime', inplace=True)
    except Exception as e:
        print(f"Error parsing dates: {str(e)}")
        print("Columns in file:", df.columns)
        return None
    
    # Display basic info
    print("\nData Overview:")
    print(f"Time Range: {df.index.min()} to {df.index.max()}")
    print(f"Number of records: {len(df)}")
    print(f"Number of missing values: {df['water_level'].isna().sum()}")
    
    # Handle missing values if any
    if df['water_level'].isna().any():
        df['water_level'] = df['water_level'].interpolate(method='time')
    
    # Plot water level data
    plt.figure(figsize=(15, 6))
    plt.plot(df['water_level'])
    plt.title(f'Water Level - {station_name}')
    plt.xlabel('Date')
    plt.ylabel('Water Level (m)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    # Resample to daily frequency (using mean) to reduce noise
    daily_data = df['water_level'].resample('D').mean()
    
    # Handle any remaining NaN values in daily data
    if daily_data.isna().any():
        print(f"Filling {daily_data.isna().sum()} NaN values in daily data")
        daily_data = daily_data.interpolate(method='time')
    
    # Check stationarity
    print("\nChecking stationarity of daily water level data:")
    is_stationary = check_stationarity(daily_data)
    
    # Plot ACF and PACF
    plt.figure(figsize=(15, 10))
    plt.subplot(211)
    plot_acf(daily_data.dropna(), ax=plt.gca(), lags=40)
    plt.subplot(212)
    plot_pacf(daily_data.dropna(), ax=plt.gca(), lags=40)
    plt.tight_layout()
    plt.show()
    
    # Split data into train and test sets
    train_size = int(len(daily_data) * 0.8)
    train, test = daily_data[:train_size], daily_data[train_size:]
    
    print(f"\nTraining set size: {len(train)}")
    print(f"Test set size: {len(test)}")
    
    # Find best ARIMA parameters
    print("\nFinding best ARIMA parameters...")
    best_params = find_best_arima_parameters(train)
    
    # Build and evaluate model
    print("\nBuilding ARIMA model with best parameters...")
    model, forecast, metrics = build_arima_model(train, test, best_params)
    
    return {
        'station_name': station_name,
        'model': model,
        'forecast': forecast,
        'metrics': metrics,
        'best_params': best_params
    }

# Main execution
def main():
    base_dir = r"c:\Users\RushenSamodya\Documents\GitHub\Final_Year_Project\labeled_data"
    
    # Path to Kelani River water level data
    kelani_dir = os.path.join(base_dir, "water_level_data", "kelani_ganga")
    
    # Check if directory exists
    if not os.path.exists(kelani_dir):
        print(f"Directory does not exist: {kelani_dir}")
        return
    
    # Get all stations for Kelani River
    stations = [f for f in os.listdir(kelani_dir) if f.endswith('.csv')]
    
    if not stations:
        print(f"No station data found in {kelani_dir}")
        return
    
    print(f"Found {len(stations)} stations for Kelani River")
    
    # Results container
    results = []
    
    # Process each station
    for station_file in stations:
        station_path = os.path.join(kelani_dir, station_file)
        station_name = station_file.replace('.csv', '')
        
        # Analyze station data
        result = analyze_water_level_station(station_path, station_name)
        
        if result:
            results.append(result)
    
    # Compare results across stations
    if results:
        print("\n\nComparison of model performance across stations:")
        comparison = pd.DataFrame([
            {
                'Station': r['station_name'],
                'Best ARIMA Order': r['best_params'],
                'RMSE': r['metrics']['rmse'],
                'MAE': r['metrics']['mae'],
                'R': r['metrics']['r2']
            }
            for r in results
        ])
        
        print(comparison)

# Execute main function
if __name__ == "__main__":
    main()
```
